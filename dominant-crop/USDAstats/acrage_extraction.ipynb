```python
filtered_df['Value'] = filtered_df['Value'].replace({'(D)': np.nan}, regex=True)  # Replace (D) with NaN
filtered_df['Value'] = filtered_df['Value'].str.replace(',', '')  # Remove commas
filtered_df['Value'] = pd.to_numeric(filtered_df['Value'], errors='coerce')  # Convert to numeric, invalid parsing will become NaN
agreggate = filtered_df.copy()
```

```python

# Function to normalize commodities
def normalize_commodity(row):
    # Define a list of commodities to normalize
    commodities_to_normalize = ['BERRIES', 'BLACKBERRIES', 'BLUEBERRIES', 'FIELD CROPS TOTALS',
       'FRUIT', 'HAY', 'STRAWBERRIES', 'APPLES', 'CHICKPEAS', 'CITRUS',
       'COTTON', 'DATES', 'FIELD CROPSS, OTHER', 'GRAPEFRUIT', 'LEMONS',
       'ORANGES', 'PEARS', 'PECANS', 'TANGERINES', 'TREE NUT', 'WHEAT',
       'CORN', 'RICE', 'SOYBEANS', 'ALMONDS', 'APRICOTS', 'AVOCADOS',
       'BARLEY', 'BEANS', 'CHERRIES', 'CITRUS, OTHER', 'COFFEE', 'FIGS',
       'GRAPES', 'HAYLAGE', 'HERBS', 'HOPS', 'MINT', 'NECTARINES', 'OATS',
       'OLIVES', 'PEACHES', 'PISTACHIOS', 'PLUMS', 'POPCORN', 'PRUNES',
       'RASPBERRIES', 'SAFFLOWER', 'SORGHUM', 'SUNFLOWER', 'TRITICALE',
       'WALNUTS', 'WILD RICE', 'EMMER & SPELT', 'MILLET', 'PEAS', 'RYE',
       'MACADAMIAS', 'BUCKWHEAT', 'CANOLA', 'HAZELNUTS', 'TOBACCO',
       'CRANBERRIES', 'LENTILS', 'FLAXSEED', 'PEANUTS',
       'VEGETABLE TOTALS', 'VEGETABLES, OTHER', 'BROCCOLI', 'CABBAGE',
       'CARROTS', 'CAULIFLOWER', 'LETTUCE', 'ONIONS', 'POTATOES',
       'SQUASH', 'CELERY', 'MELONS', 'SPINACH', 'GARLIC', 'PEPPERS',
       'TOMATOES', 'ARTICHOKES', 'SWEET CORN', 'SWEET POTATOES']
    
    # Loop through each commodity and check if it exists in the row's 'Commodity' column
    for commodity in commodities_to_normalize:
        if commodity in row['Commodity']:
            return commodity
    
    return row['Commodity']

# Apply the normalization function to the 'Commodity' column
agreggate['Commodity'] = agreggate.apply(normalize_commodity, axis=1)

# Group by relevant columns and sum the values
aggregated_df = (
    agreggate.groupby(['State', 'Commodity'], as_index=False)
    .agg({'Value': 'sum'})
)

# Pivot the table to have commodities as columns
pivot_df = aggregated_df.pivot(index='State', columns='Commodity', values='Value').reset_index()

# Fill missing values with 0 (optional, depending on data needs)
pivot_df = pivot_df.fillna(0)
# pivot_df = pivot_df.drop(columns=['FIELD CROP TOTALS','VEGETABLE TOTALS'])
# Save the result to a new CSV file
pivot_df.to_csv('aggregated_crops.csv', index=False)

print("Aggregation complete. Data saved to 'aggregated_crops.csv'.")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from matplotlib.patches import Patch

def autopct_format(pct):
    """Show percentage only if ≥ 5%"""
    return f"{pct:.1f}%" if pct >= 5 else ""

# Original fixed color palette with lowercase keys
color_palette = {
    "sorghum": "#ff9999", "soybean": "#66b3ff", "hay": "#99ff99", "flaxseed": "#ffcc99",
    "canola": "#ff6666", "safflower": "#c2c2f0", "barley": "#ffb3e6", "peanut": "#c4e17f",
    "proso_millet": "#f1b2a1", "other_spring_wheat": "#9b8f63", "dry_edible_pea": "#a3d8f4", 
    "winter_wheat": "#ed7d31", "sugarbeet": "#f3d1c1", "lentil": "#84c8b2", "rice": "#4a86e8", 
    "chickpeas": "#bc8f8f", "rye": "#d3d3d3", "cotton": "#f7b3d5", "sugarcane": "#ffb84d", 
    "oat": "#ff99cc", "tobacco": "#9c9c9c", "corn_planted": "#7db36d", "sunflower": "#ffeb3b", 
    "wheat": "#fbc02d", "dry_edible_bean": "#ab47bc", "potato": "#ff8a65", "durum_wheat": "#0d47a1"
}

# Read the CSV file (adjust the filename as needed)
df = pivot_df.copy()
df = df[df['State'] != 'OTHER STATES']

# Get a list of all commodity columns (all except 'State')
commodity_columns = df.columns.drop('State')

# Build a final color palette that covers all crops in the dataframe
final_color_palette = {}
missing = []

for crop in commodity_columns:
    norm_crop = crop.lower().replace(" ", "").replace("&", "")
    if norm_crop in color_palette:
        final_color_palette[crop] = color_palette[norm_crop]
    else:
        missing.append(crop)
if missing:
    cmap = cm.get_cmap('tab20', len(missing))
    for i, crop in enumerate(sorted(missing)):
        final_color_palette[crop] = cmap(i)

def plot_pie_charts(sub_df, fig_title, color_mapping):
    """Plot a grid of pie charts (one per state) with a single legend."""
    n_states = sub_df.shape[0]
    n_cols = min(5, n_states)
    n_rows = (n_states + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(6*n_cols, 6*n_rows))
    
    if n_states == 1:
        axes = [axes]
    else:
        axes = axes.flatten()
    
    all_labels = set()

    for i, (_, row) in enumerate(sub_df.iterrows()):
        state_name = row["State"]
        values = row.iloc[1:]
        
        if values.sum() == 0:
            axes[i].axis("off")
            continue
        
        percentages = values / values.sum() * 100
        
        labels = [col if percentages[j] >= 5 else "" 
                  for j, col in enumerate(values.index)]
        all_labels.update(values.index)
        
        pie_colors = [color_mapping[col] for col in values.index]
        
        wedges, texts, autotexts = axes[i].pie(
            values, labels=labels, autopct=lambda p: autopct_format(p), startangle=90,
            wedgeprops={"edgecolor": "black"}, radius=1.4, colors=pie_colors
        )
        
        for text in texts + autotexts:
            text.set_fontsize(10)
        
        axes[i].set_title(state_name, fontsize=12, y=1.2)
    
    for j in range(i+1, len(axes)):
        axes[j].axis("off")
    
    legend_handles = [Patch(facecolor=color_mapping[col], edgecolor='black', label=col)
                      for col in sorted(commodity_columns)]
    
    fig.legend(handles=legend_handles, loc="lower center", bbox_to_anchor=(0.5, -0.05), 
               ncol=6, fontsize=10)
    
    plt.suptitle(fig_title, fontsize=18, fontweight="bold", y=0.98)
    plt.tight_layout(rect=[0, 0.05, 1, 0.96])
    
    plt.savefig('pie_charts.png')
    plt.show()

if df.shape[0] > 25:
    df1, df2 = df.iloc[:25], df.iloc[25:]
    plot_pie_charts(df1, "% Acreage of crops grown in first 25 states", final_color_palette)
    plot_pie_charts(df2, "% Acreage of crops grown in next 25 states", final_color_palette)
else:
    plot_pie_charts(df, "% Acreage of crops grown by state", final_color_palette)
```
