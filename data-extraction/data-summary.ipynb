```python
df['What federal regulation(s) does the study assess?'] = df['What federal regulation(s) does the study assess?'].replace('Worker Protection Standard; Field Sanitation Standard', 'Both')
```

```python
import matplotlib.pyplot as plt

col = 'What federal regulation(s) does the study assess?'

counts = df[col].value_counts(dropna=False, normalize=True) * 100  # percentages
counts_df = counts.to_frame().T
counts_df.index = [""]  # remove xtick label

# Plot stacked bar
ax = counts_df.plot(
kind="bar",
stacked=True,
figsize=(8, 4),
legend=False  # turn off auto-legend, we'll add custom one
)

plt.title(col, fontsize=12, pad=20)
plt.ylabel("Percentage")
plt.xlabel("")
plt.ylim(0, 100)
plt.xticks([])

# Build custom legend with percentages
handles, _ = ax.get_legend_handles_labels()
legend_labels = [f"{cat}: {val:.1f}%" for cat, val in counts.items()]
plt.legend(handles, legend_labels, title="Category",
        bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout()

# Save as PNG (filename includes column name)
plt.savefig(f"{col.replace('/',', ')}_distribution.png", dpi=300, bbox_inches="tight")

plt.close()  # close the figure so memory doesn’t fill up
```

```python
import pandas as pd
import matplotlib.pyplot as plt

col = "What location(s) does the study cover?"

# Explode semicolon-separated values
df_exploded = (
    df[col]
    .dropna()
    .str.split(";")
    .explode()
    .str.strip()
)

# Counts
counts = df_exploded.value_counts()

# Totals
n_studies = df[col].notna().sum()     # number of studies with at least one location
n_mentions = len(df_exploded)         # total location mentions

# Bar chart
fig, ax = plt.subplots(figsize=(10, 6))
counts.plot(kind="bar", ax=ax, color="skyblue", edgecolor="black")

# Title with N
ax.set_title(
    f"Study Locations (N={n_studies} studies; {n_mentions} mentions)",
    fontsize=14, pad=20
)

# Axis labels
ax.set_xlabel("Location")
ax.set_ylabel("Count")

# Rotate x labels for readability
plt.xticks(rotation=45, ha="right")

# Explanatory note
fig.text(
    0.5, -0.05,
    "Note: Total count exceeds number of studies because one study may cover multiple locations.",
    ha="center", va="top", fontsize=9, style="italic"
)

plt.tight_layout()
plt.savefig("study_locations_barchart.png", dpi=300, bbox_inches="tight")
plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

col = "Study objectives/aims"

# Explode semicolon-separated values
df_exploded = (
    df[col]
    .dropna()
    .str.split(";")
    .explode()
    .str.strip()
)

# Get counts and percentages after exploding
counts = df_exploded.value_counts()
percentages = df_exploded.value_counts(normalize=True) * 100

summary_df = pd.DataFrame({
    "Count": counts,
    "Percentage": percentages.round(2)
})

# Get totals
n_studies = len(df[col])     # number of studies with at least one objective
n_mentions = len(df_exploded)             # total objective mentions

# Create a matplotlib figure
fig, ax = plt.subplots(figsize=(8, len(summary_df) * 0.5 + 2))
ax.axis("off")

# Add table
table = ax.table(
    cellText=summary_df.values,
    rowLabels=summary_df.index,
    colLabels=summary_df.columns,
    loc="center"
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)

# Adjust column widths
col_widths = {0: 0.3, 1: 0.3}
for (row, col_idx), cell in table.get_celld().items():
    if col_idx in col_widths:
        cell.set_width(col_widths[col_idx])

# Title with N
ax.set_title(
    f"Summary of Study Objectives/Aims (N={n_studies} studies; {n_mentions} mentions)",
    fontsize=12, pad=20
)

# Explanatory note
fig.text(
    0.5, -0.02,
    "Note: Total count exceeds number of studies because one study may have multiple objectives.",
    ha="center", va="top", fontsize=9, style="italic"
)

plt.tight_layout()
plt.savefig("Study_objectives_aims_summary_table.png", dpi=300, bbox_inches="tight")
plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

col = "What is the study’s sample size? "

# Drop NA and ensure numeric
data = pd.to_numeric(df[col], errors="coerce").dropna()

# Summary statistics
mean_val = data.mean()
std_val = data.std()
q1 = data.quantile(0.25)
q3 = data.quantile(0.75)
iqr = q3 - q1

# Create boxplot without outliers
fig, ax = plt.subplots(figsize=(6, 6))
ax.boxplot(data, vert=True, patch_artist=True, showfliers=False)

ax.set_title(col + "(N = 40 studies)", fontsize=12, pad=15)
ax.set_ylabel("Sample Size")

# Add legend-like text box with stats
stats_text = (
    f"Mean = {mean_val:.2f}\n"
    f"SD = {std_val:.2f}\n"
    f"IQR = {iqr:.2f}"
)
props = dict(boxstyle="round,pad=0.4", facecolor="lightgray", alpha=0.7)
ax.text(1.1, np.median(data), stats_text, transform=ax.get_yaxis_transform(),
        fontsize=10, verticalalignment="center", bbox=props)

plt.tight_layout()
plt.savefig("sample_size_boxplot.png", dpi=300, bbox_inches="tight")
plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Explode into separate rows for each term
df_exploded = df.assign(
    Terms_list=df["Terms Used by Author(s) in Describing the Sample"].str.split(";")
)
df_exploded["Terms_list"] = df_exploded["Terms_list"].apply(lambda x: [t.strip() for t in x])
df_exploded = df_exploded.explode("Terms_list")

# Get counts & percentages
counts = df_exploded["Terms_list"].value_counts(dropna=False)
percentages = df_exploded["Terms_list"].value_counts(dropna=False, normalize=True) * 100

n_mentions = len(df_exploded)            

# Create summary DataFrame
summary_df = pd.DataFrame({
    "Count": counts,
    "Percentage": percentages.round(2)
})

# Create a matplotlib figure
fig, ax = plt.subplots(figsize=(8, len(summary_df) * 0.35 + 1))
ax.axis("off")

# Add table
table = ax.table(
    cellText=summary_df.values,
    rowLabels=summary_df.index,
    colLabels=summary_df.columns,
    loc="center"
)

# Format table
table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1.0, 1.2)  # base scale first

# ---- Adjust column widths ----
# Specify widths (relative proportions, tweak as needed)
col_widths = [0.6, 0.2, 0.2]  # [row labels, Count, Percentage]

# Apply widths
for i, width in enumerate(col_widths):
    table.auto_set_column_width([i])
    for key, cell in table.get_celld().items():
        if key[1] == i:  # column index
            cell.set_width(width)
            
fig.text(
    0.5, -0.02,
    "Note: Total count exceeds number of studies because one study may have multiple terms.",
    ha="center", va="top", fontsize=9, style="italic"
)

plt.title(f"Summary of Terms Used by Authors, (N={n_studies} studies; {n_mentions} mentions)", fontsize=12, pad=20)

# Save as PNG
plt.tight_layout()
plt.savefig("Terms_Used_by_Authors_summary_table.png", dpi=300, bbox_inches="tight")
plt.close()
```
